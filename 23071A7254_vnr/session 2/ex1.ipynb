```python
# 1) Create a NumPy array from a list.
import numpy as np
lst=[1,2,3,4,5]
print(np.array(lst))
```

```python
# 2) Generate an array of zeros, ones, or a specific constant in NumPy.
print(np.zeros((3,3)))
print(np.ones((3,3)))
print(np.full((3,3),8))
```

```python
# 3) Find the maximum, minimum, and mean of a NumPy array.
a=np.random.randint(1,10,(3,3))
print(a)
print(np.max(a))
print(np.min(a))
print(np.mean(a))
```

```python
# 4) Stack two NumPy arrays vertically and horizontally.
a=np.array([[1,2],[3,4]])
b=np.array([[1,2],[3,4]])
print(np.vstack((a,b)))
print(np.hstack((a,b)))
```

```python
# 5) Find unique elements in a NumPy array.
a=np.array([1,2,2,3,3,4,5,5,6])
print(np.unique(a))
```

```python
# 6) Generate a random number matrix of size 5*5 using NumPy.
rand_matrix=np.random.rand(5,5)
print(rand_matrix)
```

```python
# 7) Extract an element in the matrix.
print(rand_matrix[3,4])
```

```python
# 8) Create a Pandas DataFrame from a dictionary or a list.
import pandas as pd
data_dict={'name':['x','y','z'],'age':[18,19,20]}
df1=pd.DataFrame(data_dict)
print(df1)
data_list=[['a',18],['b',19]]
df2=pd.DataFrame(data_list,columns=['name','age'])
print(df2)
```

```python
# 9) Merge or concatenate two Pandas DataFrames.
print(pd.concat([df1,df2],ignore_index=True))
```

```python
# 10) Fill missing values in the above data.
df=pd.read_csv('country_wise_latest_modified.csv')
# print(df.head())
df.fillna(df.mean(numeric_only=True),inplace=True)
print(df)
```

```python
# 11) Drop a column or a row in the above data.
print(df.columns)
df_new=df.drop(columns=['Confirmed last week'])
print(df_new)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 14) What is the relationship between confirmed cases and deaths? (Scatter plot)
plt.figure(figsize=(6,3))
sns.scatterplot(x=df['Confirmed'],y=df['Deaths'])
plt.xlabel("Confirmed cases")
plt.ylabel("Deaths")
plt.show()
```

```python
# 15) How does the number of active cases vary across different countries? (Box plot)
plt.figure(figsize=(40,5))
sns.boxplot(x=df['Country/Region'],y=df['Active'])
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Active cases")
plt.show()
```

```python
# 16) How do the number of new cases compare across different WHO regions? (Bar chart or Pie chart)
plt.figure(figsize=(10,3))
sns.barplot(x=df['WHO Region'],y=df['New cases'])
plt.xlabel("WHO Region")
plt.ylabel("New cases")
plt.show()
```

```python
plt.figure(figsize=(4, 4))
df.groupby("WHO Region")["New cases"].sum().plot.pie(autopct='%1.1f%%')
plt.ylabel("")
plt.show()
```

```python
# 17) How has the number of confirmed cases changed compared to last week? (Line plot)
plt.figure(figsize=(6,3))
sns.lineplot(x=df['1 week change'],y=df['Confirmed'])
plt.xlabel("1 week change")
plt.ylabel("Confirmed cases")
plt.show()
```
